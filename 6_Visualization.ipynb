```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_health_supplement.csv')
```

```python
df
```

## Distribution of Discount Percentage

```python

plt.figure(figsize=(10,6))
sns.histplot(df['Discount Percentage(%)'], bins=30, kde=True,color='purple')
plt.title("Distribution of Discount Percentage")
plt.xlabel("Discount (%)")
plt.ylabel("Number of Products")
plt.show()
```

##  Discount vs Rating (Scatter Plot)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount Percentage(%)',y='Rating',color='blue',data=df)

plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.title('Discount Vs Rating')
```

```python
x = df.groupby('Discount Prices(?)')['Actual Prices(?)'].mean()
```

```python
x
```

## Relation between Discount Price and Actual Price(Bar plot)

```python
plt.figure(figsize=(10,6))
plt.bar(x.index,x.values,width=70,color='green')
plt.xlabel('Discount Prices')
plt.ylabel('Actual Prices')
plt.title('Relation between Discount Price and Actual Price')
```

## Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Only use numeric columns
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="crest", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
```
