## Data Cleaning & Preprocessing

```python
import pandas as pd
df = pd.read_csv('total_health_supplement.csv')
```

```python
#lets check which are column are interger or string
df.dtypes
#shit all value which containis number data that store in number format
```

```python
# Detect Which Values Are Not Proper Numbers
import pandas as pd

# Assuming df is already loaded
cols_to_check = ['Discount Prices(?)', 'Actual Prices(?)', 'Discount Percentage(%)', 'Users']

for col in cols_to_check:
    print(f"\nChecking column: {col}")
    invalid_rows = df[col].apply(lambda x: pd.to_numeric(str(x).replace(",", "").replace("% off", "").strip(), errors='coerce')).isna()
    if invalid_rows.any():
        print(f"Non-numeric values found in column '{col}':")
        print(df.loc[invalid_rows, col])
    else:
        print("All values are numeric (after cleaning).")
```

```python
#Convert to Numeric After Cleaning
df['Discount Prices(?)'] = pd.to_numeric(df['Discount Prices(?)'].str.replace(",", "").str.strip(), errors='coerce')
df['Actual Prices(?)'] = pd.to_numeric(df['Actual Prices(?)'].str.replace(",", "").str.strip(), errors='coerce')

# Remove '% off' and convert to numeric
df['Discount Percentage(%)'] = df['Discount Percentage(%)'].str.replace('% off', '').str.strip()
df['Discount Percentage(%)'] = pd.to_numeric(df['Discount Percentage(%)'], errors='coerce')

# Remove commas in Users and convert to integer
df['Users'] = df['Users'].str.replace(",", "").str.strip()
df['Users'] = pd.to_numeric(df['Users'], errors='coerce').astype("Int64")  # Keeps NaN support
```
